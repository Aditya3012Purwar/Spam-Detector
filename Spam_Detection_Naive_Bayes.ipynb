```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df=pd.read_csv('/content/spam.csv',encoding = "ISO-8859-1")
df
```

```python
df.drop("Unnamed: 2", inplace=True, axis=1)
df.drop("Unnamed: 3", inplace=True, axis=1)
df.drop("Unnamed: 4", inplace=True, axis=1)
df.rename(columns = {'v1':'Spam', 'v2':'Text'}, inplace = True)
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
nltk.download('stopwords')
```

```python
def process_text(text):
  no_punctuation = [char for char in text if char not in string.punctuation]
  no_punctuation = ''.join(no_punctuation)
  clean_words = [word for word in no_punctuation.split() if word.lower() not in stopwords.words('english')]
  return clean_words
df['Text'].head().apply(process_text)
```

```python
messages_bow = CountVectorizer(analyzer=process_text).fit_transform(df['Text'])
X_train, X_test, y_train, y_test = train_test_split(messages_bow,df['Spam'], test_size = 0.2, random_state = 0)
```

```python
messages_bow.shape
```

```python
classifier = MultinomialNB().fit(X_train,y_train)
classifier.predict(X_train)
```

```python
y_train.values
```

```python
predict= classifier.predict(X_train)
print(classification_report(y_train, predict))
print('Confusion Matrix:\n', confusion_matrix(y_train, predict))
print('Accuracy:', accuracy_score(y_train, predict))
```

```python
predict= classifier.predict(X_test)
print(classification_report(y_test, predict))
print('Confusion Matrix:\n', confusion_matrix(y_test, predict))
print('Accuracy:', accuracy_score(y_test, predict))
```
